```python
from getdist import loadMCSamples, plots
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc_file('../fig/matplotlibrc')
```

```python
def compare_chain(fn_all, labels=['chain1', 'chain2'], colors=['red', 'blue'], contour_ls=['-', '-'], markers={'p':1.0}, fn_out=None):
    samples_all = []
    for fn in fn_all:
        samples = loadMCSamples(fn, settings={'ignore_rows':0.01})
        samples_all.append(samples)
    
    g = plots.get_subplot_plotter()
    g.settings.legend_fontsize = 30
    g.settings.axes_labelsize = 30
    g.settings.axes_fontsize = 20
    g.settings.linewidth_contour = 2.0   # 改 contour 线条宽度
    g.settings.linewidth = 2.0           # 改 1D 曲线的线宽
    g.triangle_plot(samples_all, params=['fnl_loc', 'p', 'b1', 'sn0', 'sigmas'], markers=markers, legend_labels=labels, legend_loc='upper right', filled=False, contour_colors=colors, contour_ls=contour_ls)
    if fn_out:
        plt.savefig(fn_out, bbox_inches='tight')
        print(f"Saved figure to {fn_out}")
    else:
        plt.show()
```

## test on ph001

```python
names = ['QSO_box_z3_base-A_ph000_thecov', 'QSO_box_z3_base-A_ph001_thecov', 'QSO_box_z3_base-A_ph001_thecov_fixp1.4', 'QSO_box_z3_base-A_ph001_thecov_fixp1.6']
labels=['QSO6(z=3) base-A c302ph000', 'test on ph001', 'ph001, fix p=1.4', 'ph001, fix p=1.6']
colors=['#d62728', '#1f77b4', '#2ca02c', '#ff7f0e']
contour_ls=['-', '-', '-', '-']

fn_out='plots/QSO_box_z3_base-A_thecov.png'

chains = ['works/'+name+'/chain_zeus' for name in names]
compare_chain(chains, labels=labels, markers={'fnl_loc':100}, colors=colors, contour_ls=contour_ls, fn_out=fn_out)
```
